```python
#import the pandas library
#read the laliga CSV file and store it in the dataframe df

import pandas as pd

df= pd.read_csv("Laliga.csv")
```

```python
#The column names in the dataframe df were coming as unnamed, hence making the first position rows as columns
#and creating a new data frame laliga
headers = df.iloc[0]
laliga = pd.DataFrame(df.values[1:],columns = headers)
laliga.head(10)
```

<b> Q-1: Read the data set and replace dashes with 0 to make sure you can perform arithmetic operations on the data.</b>

```python
#use the replace function by giving which values to be replaced first followed by the value
#Below we are replacing the '-' with 0
laliga = laliga.replace('-',0)
laliga
```

<b>Q2:Print all the teams which have started playing between 1930-1980</b>

```python
#taking the first four charaters of Debut column by using the str function on that column as we need just debut year
#Now compare the Debut column if >= 1930 or <= 1980 to get the desired output and store that in lalaiga_1 dataframe
#Using the comparision in quotes as the Debut column is a string

laliga_1 = laliga.loc[((laliga.Debut.str[0:4]) >= '1930') & ((laliga.Debut.str[0:4]) <= '1980')]

#Displaying the Team and Debut year from the new dataframe laliga_1
laliga_1.loc[:,['Team','Debut']]
```

<b>Q3:Print the list of teams which came Top 5 in terms of points</b>

```python
#Sort the dataframe so that the maximum points come on top
laliga2 = laliga.sort_index()

#print the top 5 teams with points
laliga2.loc[:4,('Team','Points')]
```

<b>Q4:Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences.
      Using the same function,find the team which has maximum and minimum goal difference.</b>

```python
#Define a function "Goal_diff_count" and pass two values GoalsFor and GoalsAgainst, which returns the difference of the two.

def Goal_diff_count(GoalsFor,GoalsAgainst):
    Goal_diff_count = GoalsFor - GoalsAgainst
    return Goal_diff_count
```

```python
#Create a new column "Goal Difference" which calls the function Goal_diff_count for GoalsFor and GoalsAgainst values
#and returns the Goal diff count and stores that in the new column "Goal Difference" of the laliga Dataframe

laliga['Goal Difference'] = Goal_diff_count((pd.to_numeric(laliga["GoalsFor"])),pd.to_numeric(laliga["GoalsAgainst"]))
laliga.head(5)
```

```python
#Display the Team along with the Goal Difference

laliga[['Team','Goal Difference']]
```

```python
#Displays the Team name along with Goal difference which has highest Goal Difference
laliga[['Team','Goal Difference']][laliga['Goal Difference'] == laliga['Goal Difference'].max()]
```

```python
#Displays the Team name along with Goal difference which has lowest Goal Difference
laliga[['Team','Goal Difference']][laliga['Goal Difference'] == laliga['Goal Difference'].min()]
```

<b>Q:5 - Create a new column with name “Winning Percent” and append it to the data set</b>

<b>If there are any numerical error, replace it with 0%</b>

<b>Print the top 5 teams which has the highest Winning percentage</b>

```python
#First Convert the columns required for calculation into numeric, "GamesWon" and "GamesPlayed"

laliga['GamesWon'] = pd.to_numeric(laliga['GamesWon'])
laliga['GamesPlayed'] = pd.to_numeric(laliga['GamesPlayed'])
```

```python
#Calculate the winning percentage by dividing GamesWon and GamesPlayed
#Append the results to the Dataframe in the column "Winning Percent"
#The last row of winning percentage was produced as Nan

laliga['Winning Percent'] = (laliga['GamesWon']).div(laliga['GamesPlayed'])

laliga.tail(5)
```

```python
#Since we get a NaN when we divide by zero, hence we are using Numpy to replace the same value with 0
#after the replace it is converted to 0

import numpy as np

laliga['Winning Percent'] = (laliga['GamesWon']).div(laliga['GamesPlayed'],fill_value=0).replace({np.NaN:0}) * 100

laliga.tail(5)
```

```python
#Display the Team along with its Winning percent
laliga.loc[:,('Team','Winning Percent')]
```

```python
#top 5 teams with highest winning percentage
#sort by winning percentage then select top 5

laliga4 = laliga.sort_values(by=['Winning Percent'],ascending=False)

laliga4.loc[:4,('Team','Winning Percent')]
```

<b>Q6 :Group teams based on their “Best position” and print the sum of their points for all positions</b>

```python
#Conver the points column to Numeric
#Select the columns "Best position" and "Points" with which we wanted to group
#Store that into a dataframe laliga5

laliga['Points'] = pd.to_numeric(laliga['Points'])

laliga5 = laliga[['BestPosition','Points']]
laliga5
```

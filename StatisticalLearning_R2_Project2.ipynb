### Q2.Use the five percent significance test over the data to determine the p value to 
###       check new scheme has significantly raised outputs? 

```python
#Import the stats libraries for ttest

from scipy.stats import ttest_1samp,ttest_ind,levene,shapiro 
from statsmodels.stats.power import ttest_power
```

```python
#For the above question our hypothesis is as below
#Ho :µ_new_scheme - µ_old_Scheme = 0 (Null hypothesis) or µ_new_Scheme = µ_old_scheme - No raise
#Ha :µ_new_scheme - µ_old_Scheme > 0 (Alternative hypothesis) or µ_new_Scheme > µ_old_scheme - There is raise
```

```python
#Divide the data into two groups of  old scheme and new scheme

old_scheme = titan_insurance.iloc[:, 1]
new_scheme = titan_insurance.iloc[:, 2]
```

```python
#first check the Normality disturbution of the samples using shapiro
#Null hypothesis Distribution is Normal
shapiro(new_scheme)
```

```python
#In the above shapiro test ,the p value is > 0.05 Do not reject Ho(null hypothesis).
#Which means the data is normally distrubuted.
```

```python
#We can go for the paired t-test 
t_statistic, p_value = ttest_1samp(new_scheme-old_scheme,0)
print(t_statistic, p_value)

#This is a two tailed test, but what we wanted to check is whether the new scheme has raised ouputs
#which is a one tailed test, hence we divide the value of p by 2

print('Real value of p: {}'.format(p_value/2))
```

```python
# p > 0.05 => Null hypothesis:
# There is no significant raise in the output in the new scheme.
print ("paired t-test p-value=", p_value)
```

### Q3.What conclusion does the test (p-value) lead to? (2.5 points)

```python
#With the p value being > 0.05 we say that the Null hypothesis holds good. i.e. #Ho :µ1 - µ2 = 0
#We have enough evidance to prove that the new scheme has no significant raise in the output.
#We are 95% sure that the new scheme has no significant raise in the output.
```
